```python
import pandas as pd
import matplotlib.pyplot as plt
import descartes
import geopandas as gpd
from shapely.geometry import Point, Polygon

%matplotlib inline
```

```python
ausMap = gpd.read_file('aust_cd66states.shx')
```

```python
fig,ax = plt.subplots(figsize = (15,15))
ausMap.plot(ax = ax)
```

```python
d1 = pd.read_csv('MajorAviationTerminals_v1.csv')
crs = {'init': 'espg:4326'}
d1.head()
```

```python
d2 = pd.read_csv('AviationControlTowers_v1.csv')
crs = {'init': 'espg:4326'}
d2.head()
```

```python
d3 = pd.read_csv('MajorAviationFuelDepots_v1.csv')
crs = {'init': 'espg:4326'}
d3.head()
```

```python
d4 = pd.read_csv('MajorAviationHangars_v1.csv')
crs = {'init': 'espg:4326'}
d4.head()
```

```python
geometry = [Point(xy) for xy in zip( d1["LONGITUDE"], d1["LATITUDE"])]
geometry[:3]
```

```python
geo_d1 = gpd.GeoDataFrame(d1, crs=crs, geometry = geometry)
geo_d1.head()
```

```python
fig,ax = plt.subplots(figsize = (15,15))
ausMap.plot(ax = ax, alpha = 0.4, color="black")
geo_d1.plot(ax = ax, markersize = 40, color = "red", marker = "x", label = "Terminal")

plt.legend(prop={'size': 15})
```

```python
geometry2 = [Point(xy) for xy in zip( d2["LONGITUDE"], d2["LATITUDE"])]
geometry2[:3]
```

```python
geo_d2 = gpd.GeoDataFrame(d2, crs=crs, geometry = geometry2)
geo_d2.head()
```

```python
fig,ax = plt.subplots(figsize = (15,15))
ausMap.plot(ax = ax, alpha = 0.4, color="black")
geo_d2.plot(ax = ax, markersize = 40, color = "blue", marker = "x", label = "Cont. Tower")

plt.legend(prop={'size': 15})
```

```python
geometry3 = [Point(xy) for xy in zip( d3["LONGITUDE"], d3["LATITUDE"])]
geometry3[:3]
```

```python
geo_d3 = gpd.GeoDataFrame(d3, crs=crs, geometry = geometry3)
geo_d3.head()
```

```python
fig,ax = plt.subplots(figsize = (15,15))
ausMap.plot(ax = ax, alpha = 0.4, color="black")
geo_d3.plot(ax = ax, markersize = 40, color = "orange", marker = "x", label = "Fuel Depot")

plt.legend(prop={'size': 15})
```

```python
geometry4 = [Point(xy) for xy in zip( d4["LONGITUDE"], d4["LATITUDE"])]
geometry4[:3]
```

```python
geo_d4 = gpd.GeoDataFrame(d4, crs=crs, geometry = geometry4)
geo_d4.head()
```

```python
fig,ax = plt.subplots(figsize = (15,15))
ausMap.plot(ax = ax, alpha = 0.4, color="black")
geo_d4.plot(ax = ax, markersize = 40, color = "green", marker = "x", label = "Hangar")

plt.legend(prop={'size': 15})
```

```python
fig,ax = plt.subplots(figsize = (15,15))
ausMap.plot(ax = ax, alpha = 0.4, color="black")
geo_d2.plot(ax = ax, markersize = 60, color = "blue", marker = "o", label = "Cont. Tower")
geo_d1.plot(ax = ax, markersize = 35, color = "red", marker = "o", label = "Terminal")
geo_d3.plot(ax = ax, markersize = 25, color = "orange", marker = "o", label = "Fuel Depot")
geo_d4.plot(ax = ax, markersize = 5, color = "green", marker = "o", label = "Hangar")

plt.legend(prop={'size': 15})
```
